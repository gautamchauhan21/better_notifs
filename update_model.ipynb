```python
def update_dumps(user_id, parent_id, category_id, val) :
    if val == 1 :
        sql = "update dumps set {0} = {1} + 0.02 where userid={2}"
        cat = str(parent_id) + "_" + str(category_id)
        print(cat)
        sql = sql.format(cat, cat, user_id)
        cursor.execute(sql)
    if val == 0 :
        sql = "update dumps set {0} = {1} - 0.01 where userid={2}"
        cat = str(parent_id) + "_" + str(category_id)
        print(cat)
        sql = sql.format(cat, cat, user_id)
        cursor.execute(sql)
    sql2 = "update receives set created_at = '1' where user_id = " + str(user_id);
    cursor.execute(sql2)
    conn.commit()
```
